```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('./Data/lena.png')
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_RGB)
```

```python
img_rec = img_RGB.copy()
cv2.rectangle(img_rec, (60, 40), (420, 510), (0, 0, 255), 2)
plt.imshow(img_rec)
```

```python
img_fill = img_RGB.copy()
cv2.rectangle(img_fill, (60, 40), (420, 510), (0, 0, 255), -1)
plt.imshow(img_fill)
```

```python
img_line = img_RGB.copy()
cv2.line(img_line, (60,40), (400, 480), (0, 255, 0), 3)
plt.imshow(img_line)
```
